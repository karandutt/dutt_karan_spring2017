```python
from collections import Counter 
count = nltk.corpus.gutenberg.words('austen-persuasion.txt')
counts = Counter(count)
print(counts)
```

```python
from collections import Counter 
count = nltk.corpus.gutenberg.words('austen-persuasion.txt')
counts = Counter(count)
most_frequent = counts.most_common(10)
#print(most_frequent)
for word, count in most_frequent:
    print('The word/special character ' + word + ' has a frequency of : ' + str(count))
```
