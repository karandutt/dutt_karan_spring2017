```python
import nltk
```

```python
#file names under package gutenberg
nltk.corpus.gutenberg.fileids()
```

```python
#finds how many words in the file austen-persuasion.txt
nltk.corpus.gutenberg.fileids()
list = nltk.corpus.gutenberg.fileids() #get all the file names and put them in a list
count = nltk.corpus.gutenberg.words('austen-persuasion.txt') #get the words from the txt file and put them in a list
print('The number of words in the file ' + list[1] + ' is : ' + str(len(count))) #list[1] gives austen-persuasion 
#and i converted int to string to display it in text fromat
```

```python
#finds how many unique words in the file austen-persuasion.txt
nltk.corpus.gutenberg.fileids()
list = nltk.corpus.gutenberg.fileids() #get all the file names and put them in a list
count = nltk.corpus.gutenberg.words('austen-persuasion.txt') #get the words from the txt file and put them in a list
print('The number of unique words in the file ' + list[1] + ' is : ' + str(len(set(count)))) 
#set function basically removes duplicates
```

```python
#calculate the frequency of each word in austen-persuasion.txt
text = nltk.corpus.gutenberg.words('austen-persuasion.txt')
#print(text)
Dict = {} #create an empty dictionary
for word in text: #for each word in the austen-persuasion text loop
   Dict[word]= Dict.get(word, 0) + 1 #key : word, value of the key is the number of occurences. replaced by default
print(Dict)
```

```python
#display the top 10 words used in austen-persuasion.txt
count = nltk.corpus.gutenberg.words('austen-persuasion.txt')
Dict = {}
for word in count:
    if word in Dict:
        Dict[word] += 1 #if the word is already present +1
    else:
        Dict[word] = 1 # found a new word so count starts from 1
top = sorted(Dict, key = Dict.get, reverse = True) #reverse = False gives least common, surfed stackoverflow for this line
top_10 = top[:10] #using the slice operator from 0 to 9 gives top 10 values which have been sorted  
print(top_10) #print the values
 
```
