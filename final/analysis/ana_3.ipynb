# Analysis 3 : 
## Visualize temperature in India over time

```python
#do all the imports
import pandas as pd
import seaborn as sns
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from pandas import set_option
#set option allows you to control how pandas will display your results
#reduced the amount of results to be displayed by the following code
set_option("display.max_rows",16)
LARGE_FIGSIZE = (12,8)
```

```python
#use pandas function read_csv to read the csv file
global_temp_country = pd.read_csv("GlobalLandTemperaturesByCountry.csv")
#use head or tail to get few rows of data from the DataFrame
global_temp_country.tail()
```

```python
global_temp_country.describe()
```

```python
global_temp_country["AverageTemperatureUncertainty"].plot()
plt.savefig("./ana_3/tempuncplot.pdf", bbox_inches='tight')
```

```python
#kernel density estimate plot using seaborn
global_temp_country["AverageTemperatureUncertainty"].plot(kind = "kde")
plt.savefig("./ana_3/tempunckdeplot.pdf", bbox_inches='tight')
```

```python
country_india = global_temp_country[global_temp_country.Country == 'India']
country_india.set_index(['dt'], inplace=True)
country_india
```

```python
india_avg = country_india.AverageTemperature
```

```python
#for India, plot both the Average Temp and Average Temp Uncertainity from the year 1796-2004
country_india.plot(figsize=(10, 8))
plt.savefig("./ana_3/avgtempavgtempuncertainty.pdf", bbox_inches='tight')
```

```python
india_avg.hist(bins=100, figsize=(10, 8))
plt.savefig("./ana_3/indiaavghist.pdf", bbox_inches='tight')
```

```python
from pandas.tools import plotting
plotting.scatter_matrix(country_india, alpha=0.8, figsize=(10, 8), diagonal='kde')
plt.savefig("./ana_3/indiaavgscattertempunc.pdf", bbox_inches='tight')
```
